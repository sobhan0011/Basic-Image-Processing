# Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Loading original pictures

```python
image = cv2.imread('girlface.bmp', cv2.IMREAD_GRAYSCALE)
original_shape = image.shape
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```

# Functions to extract and show the bit planes

```python
def extract_bit_plane(image, bit):
    bit_to_extract = bit
    binary_image = np.unpackbits(image)
    bit_plane = binary_image[bit::8]
    bit_plane = np.reshape(bit_plane, (image.shape[0], image.shape[1]))
    bit_plane = bit_plane.astype(np.uint8) * 255
    return bit_plane    
```

```python
bit_7_plane = extract_bit_plane(image, 0)
bit_6_plane = extract_bit_plane(image, 1)
bit_5_plane = extract_bit_plane(image, 2)
bit_4_plane = extract_bit_plane(image, 3)
bit_3_plane = extract_bit_plane(image, 4)
bit_2_plane = extract_bit_plane(image, 5)
bit_1_plane = extract_bit_plane(image, 6)
bit_0_plane = extract_bit_plane(image, 7)
```

```python
fig, axes = plt.subplots(2, 4, figsize=(12, 6))
axes[0, 0].imshow(bit_7_plane, cmap='gray', vmin=0, vmax=255)
axes[0, 0].axis('off')
axes[0, 0].set_title(f'Bit Plane 8')

axes[0, 1].imshow(bit_6_plane, cmap='gray', vmin=0, vmax=255)
axes[0, 1].axis('off')
axes[0, 1].set_title(f'Bit Plane 7')

axes[0, 2].imshow(bit_5_plane, cmap='gray', vmin=0, vmax=255)
axes[0, 2].axis('off')
axes[0, 2].set_title(f'Bit Plane 6')

axes[0, 3].imshow(bit_4_plane, cmap='gray', vmin=0, vmax=255)
axes[0, 3].axis('off')
axes[0, 3].set_title(f'Bit Plane 5')



axes[1, 0].imshow(bit_3_plane, cmap='gray', vmin=0, vmax=255)
axes[1, 0].axis('off')
axes[1, 0].set_title(f'Bit Plane 4')

axes[1, 1].imshow(bit_2_plane, cmap='gray', vmin=0, vmax=255)
axes[1, 1].axis('off')
axes[1, 1].set_title(f'Bit Plane 3')

axes[1, 2].imshow(bit_1_plane, cmap='gray', vmin=0, vmax=255)
axes[1, 2].axis('off')
axes[1, 2].set_title(f'Bit Plane 2')

axes[1, 3].imshow(bit_0_plane, cmap='gray', vmin=0, vmax=255)
axes[1, 3].axis('off')
axes[1, 3].set_title(f'Bit Plane 1')

plt.tight_layout()
plt.show()
```

```python
def make_one_plane_zero(image, bit):
    binary_image = np.unpackbits(image)
    binary_image[bit::8] = 0
    repacked_image = np.packbits(binary_image).reshape(image.shape)
    return repacked_image
```

```python
exclude_bit_7_plane = make_one_plane_zero(image, 0)
exclude_bit_6_plane = make_one_plane_zero(image, 1)
exclude_bit_5_plane = make_one_plane_zero(image, 2)
exclude_bit_4_plane = make_one_plane_zero(image, 3)
exclude_bit_3_plane = make_one_plane_zero(image, 4)
exclude_bit_2_plane = make_one_plane_zero(image, 5)
exclude_bit_1_plane = make_one_plane_zero(image, 6)
exclude_bit_0_plane = make_one_plane_zero(image, 7)
```

```python
fig, axes = plt.subplots(2, 4, figsize=(12, 6))
axes[0, 0].imshow(exclude_bit_7_plane, cmap='gray', vmin=0, vmax=255)
axes[0, 0].axis('off')
axes[0, 0].set_title(f'Exclude Bit Plane 8')

axes[0, 1].imshow(exclude_bit_6_plane, cmap='gray', vmin=0, vmax=255)
axes[0, 1].axis('off')
axes[0, 1].set_title(f'Exclude Bit Plane 7')

axes[0, 2].imshow(exclude_bit_5_plane, cmap='gray', vmin=0, vmax=255)
axes[0, 2].axis('off')
axes[0, 2].set_title(f'Exclude Bit Plane 6')

axes[0, 3].imshow(exclude_bit_4_plane, cmap='gray', vmin=0, vmax=255)
axes[0, 3].axis('off')
axes[0, 3].set_title(f'Exclude Bit Plane 5')



axes[1, 0].imshow(exclude_bit_3_plane, cmap='gray', vmin=0, vmax=255)
axes[1, 0].axis('off')
axes[1, 0].set_title(f'Exclude Bit Plane 4')

axes[1, 1].imshow(exclude_bit_2_plane, cmap='gray', vmin=0, vmax=255)
axes[1, 1].axis('off')
axes[1, 1].set_title(f'Exclude Bit Plane 3')

axes[1, 2].imshow(exclude_bit_1_plane, cmap='gray', vmin=0, vmax=255)
axes[1, 2].axis('off')
axes[1, 2].set_title(f'Exclude Bit Plane 2')

axes[1, 3].imshow(exclude_bit_0_plane, cmap='gray', vmin=0, vmax=255)
axes[1, 3].axis('off')
axes[1, 3].set_title(f'Exclude Bit Plane 1')

plt.tight_layout()
plt.show()
```
